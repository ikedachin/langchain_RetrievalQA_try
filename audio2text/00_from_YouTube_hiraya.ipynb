```python
import os
import time
import whisper
```

## モデルのダウンロードと、audioデータの取り出し

```python
model_size = 'medium' # 'large', 'medium', 'small', 'base'
model = whisper.load_model(model_size) # Install Certificates.commandをダブルクリックして更新
```

# 開始時間の記録

```python
start_time = time.time()
```

# オーディオデータの読み込み

```python
audio = whisper.load_audio('../audio_files/【平屋建築④】職人社長が本気で家をつくる！驚きの性能やその裏側をぜんぶ見せます！【注文住宅 外壁 新築】.mp4')
transcription = model.transcribe('../audio_files/【平屋建築④】職人社長が本気で家をつくる！驚きの性能やその裏側をぜんぶ見せます！【注文住宅 外壁 新築】.mp4')
transcription
```

```python
with open(f'../text_files/hiraya_whisper_{model_size}.txt', 'w') as file:
    # Write each item in the list to the file
    for i in range(len(transcription['segments'])):
        file.write(str(transcription['segments'][i]['text']) + "\n")
```

```python
from datetime import datetime
import time

texts = []
start_times = []

for segment in transcription['segments']:
    text = segment['text']
    start = segment['start']
    # print(start)
    start_datetime = datetime.fromtimestamp(start)
    # print(start_datetime)
    formatted_start_time = start_datetime.strftime('00:%M:%S')
    print(formatted_start_time)

    # texts.append("".join(text))
    texts.append(text)
    start_times.append(formatted_start_time)

print(texts[:10])
print(start_times[:10])
```
